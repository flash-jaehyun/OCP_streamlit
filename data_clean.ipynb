```python
import pandas as pd
```

```python
df_H = pd.read_csv('https://raw.githubusercontent.com/flash-jaehyun/OCP_streamlit/main/data/H_raw_data.csv')
df_OH = pd.read_csv('https://raw.githubusercontent.com/flash-jaehyun/OCP_streamlit/main/data/OH_raw_data.csv')
df_O = pd.read_csv('https://raw.githubusercontent.com/flash-jaehyun/OCP_streamlit/main/data/O_raw_data.csv')
```

```python
df_H['bulk_mpid'] = df_H['bulk_mpid'].astype('str') 
df_OH['bulk_mpid'] = df_OH['bulk_mpid'].astype('str') 
df_O['bulk_mpid'] = df_O['bulk_mpid'].astype('str') 

df_H['bulk_sybmols'] = df_H['bulk_symbols'].astype('str') 
df_H['bulk_sybmols'] = df_H['bulk_symbols'].astype('str') 
df_H['bulk_sybmols'] = df_H['bulk_symbols'].astype('str') 
df_OH['bulk_mpid'] = df_OH['bulk_mpid'].astype('str') 
df_O['bulk_mpid'] = df_O['bulk_mpid'].astype('str') 

df_H['bulk_mpid'] = df_H['bulk_mpid'].astype('str') 
df_OH['bulk_mpid'] = df_OH['bulk_mpid'].astype('str') 
df_O['bulk_mpid'] = df_O['bulk_mpid'].astype('str') 



bulk_mpids = []
bulk_symbols = []
ads_E_Hs = []
ads_E_Os = []
ads_E_OHs = []

for i in range(len(df_OH)):
    bulk_mpid = (df_OH.loc[i,['bulk_mpid']])
    bulk_symbol = df_OH.loc[i,['bulk_symbols']]
    ads_E_OH = (df_OH.loc[i,['adsorption_energy']])
    #miller = df_OH.loc[i, ['miller_index']]
    
    #print(miller)
    #istop = df_OH.loc[i, ['top']]
    match_df_H = df_H[df_H['bulk_mpid'].isin(bulk_mpid)] 
    match_df_O = df_O[df_O['bulk_mpid'].isin(bulk_mpid)] 
    #print(match_df_O.empty)

    #if not match_df_O.empty and not match_df_H.empty and (not bulk_mpid in bulk_mpids):
    if (match_df_O.empty is False) and (match_df_H.empty is False): #and (bulk_mpid in bulk_mpids is False): 
        bulk_mpids.append(str(df_OH.loc[i,['bulk_mpid']]).split()[1])
        bulk_symbols.append(str(df_OH.loc[i,['bulk_symbols']]).split()[1])

        match_df_H = match_df_H.reset_index()
        ads_E_Hs.append(float(match_df_H.loc[0,['adsorption_energy']]))
        #print(float(match_df_H.loc[0,['adsorption_energy']]))

        match_df_O = df_O[df_O['bulk_mpid'].isin(bulk_mpid)] 
        match_df_O = match_df_O.reset_index()
        ads_E_Os.append(float(match_df_O.loc[0,['adsorption_energy']]))
        ads_E_OHs.append(str(df_OH.loc[i,['adsorption_energy']]).split()[1])
    
```

```python
final_df = pd.DataFrame(
    {
        "bulk_mpid" : bulk_mpids,
        "bulk_symbol" : bulk_symbols,
        "ads_E_H" : ads_E_Hs,
        "ads_E_O" : ads_E_Os,
        "ads_E_OH" : ads_E_OHs,
        #"miller_index" : [],
        #"elements" : [],
        #"comparable" : []
    }
)
```
